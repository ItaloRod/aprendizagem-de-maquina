```python
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
X,y = datasets.load_iris(return_X_y=True)
X = MinMaxScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X,y)
arr_hidden_layer_sizes = range(10,101,10)
```

```python
mlp_train = []
mlp_test = []
for size in arr_hidden_layer_sizes:
    mlp = MLPClassifier(hidden_layer_sizes=size)
    mlp.fit(X_train, y_train)
    mlp_train.append(mlp.score(X_train, y_train))

for size in arr_hidden_layer_sizes:
    mlp = MLPClassifier(hidden_layer_sizes=size)
    mlp.fit(X_train, y_train)
    mlp_test.append(mlp.score(X_test, y_test))

```

```python
plt.plot(arr_hidden_layer_sizes, mlp_train)
plt.show()
```

```python
plt.plot(arr_hidden_layer_sizes, mlp_test)
plt.show()
```
