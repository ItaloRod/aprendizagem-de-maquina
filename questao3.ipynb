```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = load_breast_cancer(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
scores_train = []
scores_test = []
c_range = np.arange(0.1, 1, 0.1)
```

```python
for c in c_range:
    svm = SVC(kernel='linear',C=c,degree=2)
    svm.fit(X_train,y_train)
    scores_train.append(svm.score(X_train,y_train))
```

```python
for c in c_range:
    svm = SVC(kernel='linear',C=c,degree=2)
    svm.fit(X_train,y_train)
    scores_test.append(svm.score(X_test,y_test))
```

```python
# Gráfico, 
plt.scatter(c_range, scores_train)
plt.show()
```

```python
plt.scatter(c_range, scores_test)
plt.show()
```
