```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit
```

```python
df = pd.read_csv('ex1data1.txt', delimiter=',', header=None, names=['x','y'])
df.head(10)
```

```python
X = df['x']
y = df['y']
```

```python
plt.scatter(X,y)
plt.show()
```

```python
# Regress√£o linear aplicada
b, m = polyfit(X, y, 1)
plt.plot(X, y, '.')
plt.plot(X, b + m * X, '-')
plt.show()
```
